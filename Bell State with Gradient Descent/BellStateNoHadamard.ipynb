# $|\phi_{+}>$ Bell State Without Using a Hadamard Gate

Note: For the full solution where Gradient Descent is used to obtain the $R_x$ and $R_y$ parameters, check out Cohort2Task2.ipynb

```python
from qiskit import *
import numpy as np
PI = np.pi
```

```python
qc = QuantumCircuit(2,2)
qc.rx(PI/2,0)
qc.ry(-PI,0)
qc.cnot(0,1)
qc.draw('mpl')
```

```python
qc.measure(0,0)
qc.measure(1,1)

simulator = Aer.get_backend('qasm_simulator')

result = execute(qc, backend = simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram

plot_histogram(result.get_counts(qc))
```
