```python
from qiskit import *
import numpy as np
PI = np.pi

#See GradDescBellState.ipynb for full solution and explaination!
```

```python
qc = QuantumCircuit(2,2)
```

```python
qc.rx(PI/2,0)
qc.ry(-PI,0)
qc.cnot(0,1)
qc.draw('mpl')
```

```python
qc.measure(0,0)
qc.measure(1,1)

simulator = Aer.get_backend('qasm_simulator')

result = execute(qc, backend = simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram

plot_histogram(result.get_counts(qc))
```
